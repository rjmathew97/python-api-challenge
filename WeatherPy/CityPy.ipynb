```python
# list of lat/lon combination 
lat_lngs = []

# list of cities from citipy
cities = [] 

# specify ranges of lat/lng values
lat_range = (-90, 90)
lng_range = (-180, 180)

#generate sets of random latitude and longitudes 
lats = np.random.uniform(lat_range[0], lat_range[1], size=3000) # generate 10 latitudes
lngs = np.random.uniform(lng_range[0], lng_range[1], size=3000) # generate 1 longitudes

print(lats)
print(lngs)

#use zip to combine the lats and lons into pairs 
coordinateList = list(zip(lats, lngs))

for coord in coordinateList:
    city = citipy.nearest_city(coord[0], coord[1]).city_name
    if city not in cities:
        cities.append(city)
    print(coord[0], coord[1])
          
print(f"Number of unique cities found: {len(cities)}")
```
