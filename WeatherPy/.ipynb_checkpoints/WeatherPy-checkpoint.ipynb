## Starter Code to Generate Random Geographic Coordinates and a List of Cities

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
from scipy.stats import linregress
from pprint import pprint

# Impor the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy

print(f"API Key: {weather_api_key}")
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = list(zip(lats, lngs))
#print(lats)
#print(lags)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name

    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f"Number of cities in the list: {len(cities)}")
```

## Requirement 1: Create Plots to Showcase the Relationship Between Weather Variables and Latitude

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
# Set the API base URL
url = "http://api.openweathermap.org/data/2.5/weather"

# parameters for the API call
params = {
    "appid": weather_api_key,
    "units": "metric"
}

# Define an empty list to fetch the weather data for each city
city_data = []

# Print to logger
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all the cities in our list to fetch weather data
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0

    # Create endpoint URL with each city
    params["q"] = city
    city_url = f"{url}?q={city}&appid={params['appid']}&units={params['units']}"
    # print (city_url)

    # Log the url, record, and set numbers
    print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        response = requests.get(city_url, params=params)
        response.raise_for_status()
        response = response.json()
    

        # Parse out latitude, longitude, max temp, humidity, cloudiness, wind speed, country, and date
        city_lat = response["coord"]["lat"]
        city_lng = response["coord"]["lon"]
        city_max_temp = response["main"]["temp_max"]
        city_humidity = response["main"]["humidity"]
        city_clouds = response["clouds"]["all"]
        city_wind = response["wind"]["speed"]
        city_country = response["sys"]["country"]
        city_date = response["dt"]

        # Append the City information into city_data list
        city_data.append({"City": city,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
        

    # If an error is experienced, skip the city
    except Exception as e:
        print("City not found. Skipping... Error: {e}")
        pass
        
    # pause to avoid rate limiting
    time.sleep(1)

# Indicate that Data Loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Define a function to create Linear Regression plots
def plot_linear_regression(x, y, hemisphere, variable, xlabel, ylabel, filename):
    """
    Creates a scatter plot with a linear regression line.
    
    Parameters:
     x (array-like): Independent variable (e.g., Latitude)
    y (array-like): Dependent variable (e.g., Temperature)
    hemisphere (str): 'Northern' or 'Southern' Hemisphere
    variable (str): Variable being analyzed (e.g., 'Temperature', 'Humidity')
    xlabel (str): Label for the x-axis
    ylabel (str): Label for the y-axis
    filename (str): Name of the file to save the plot
    """
    
     # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    regression_line = slope * x + intercept

    # Create the scatter plot
    plt.scatter(x, y, edgecolor="black", alpha=0.75, label="Data Points")
    plt.plot(x, regression_line, color="red", label="Regression Line")
    plt.title(f"{hemisphere} Hemisphere: {variable} vs. Latitude")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

     # Annotate the regression formula and r^2 value
    plt.annotate(f"y = {slope:.2f}x + {intercept:.2f}", xy=(0.05, 0.90), xycoords="axes fraction", fontsize=12, color="red")
    plt.annotate(f"r^2 = {r_value**2:.2f}", xy=(0.05, 0.85), xycoords="axes fraction", fontsize=12, color="red")
    
    # Display the plot
    os.makedirs("output_data", exist_ok=True)
    plt.savefig(f"output_data/{filename}.png")
    plt.show()
```

###  Temperature vs. Latitude Linear Regression Plot

```python
#Linear Regression on North Hemisphere

#Define x and y labels 
x = np.array(northern_hemi_df["Lat"])
y = np.array(northern_hemi_df["Max Temp"])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
regression_line = slope * x + intercept

# Generate the regression equation and r^2 value
lineEq = f"y = {slope:.2f}x + {intercept:.2f}"
r_squared = f"r^2 = {r_value**2:.2f}"
print(lineEq)
print(r_squared)

# Create scatter plot with regression line
plt.scatter(x, y, color="blue", alpha=0.7, edgecolor="black")
plt.plot(x, regression_line, color="red")

# Annotate the regression equation and r^2 value
plt.annotate(lineEq, xy=(x.min() + 5, y.min() + 5), color="red")
plt.annotate(r_squared, xy=(x.min() + 5, y.min() + 10))

# Display plot
plt.title("Northern Hemisphere: Latitude vs Max Temp")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (Â°C)")
plt.show()
```

```python
# Linear regression on Southern Hemisphere
southern_hemi_df = city_data_df[city_data_df["Lat"] < 0]

# Ensure there are no missing values
southern_hemi_df = southern_hemi_df.dropna(subset=["Lat", "Max Temp"])

# Define x (Latitude) and y (Max Temperature)
x = np.array(southern_hemi_df["Lat"])
y = np.array(southern_hemi_df["Max Temp"])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
regression_line = slope * x + intercept

# Generate the regression equation and r^2 value
lineEq = f"y = {slope:.2f}x + {intercept:.2f}"
r_squared = f"r^2 = {r_value**2:.2f}"
print(lineEq)
print(r_squared)

# Plot scatter plot and regression line for Southern Hemisphere
plt.scatter(x, y, color="blue", alpha=0.75, label="Data Points")
plt.plot(x, regression_line, color="red", label="Regression Line")
plt.annotate(lineEq, xy=(min(x) + 10, max(y) - 5), fontsize=12, color="red")
plt.annotate(r_squared, xy=(min(x) + 10, max(y) - 10), fontsize=12, color="red")
plt.title("Southern Hemisphere: Latitude vs Max Temp")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (Â°C)")
plt.show()
```

```python
## Discussion about the linear relationship: Latitude vs Max Temp: The linear regression models protray a strong negative correlation 
between latitude and maximum temperature in the Northern Hemisphere; whereas, there is a moderate positive correlation in the Southern 
Hemisphere. In the Northern Hemisphere, temperatures decrease significantly as latitude increases, while in the Southern Hemisphere, 
temperature increases as latitude gets closer to the equator. Since the northern hemisphere has a stronger correlation, it suggests that 
latitude is a dominant factor in temperature variability due to its larger landmass. There is probably more oceanic influence in the 
southern hemisphere that may moderate the temperature-latitude correlation. 
```
